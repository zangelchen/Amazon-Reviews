```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import json
```

```python
music_list=[]
filepath_music = os.path.join("Digital_Music_5.json")
with open(filepath_music) as jsonfile:
    for line in jsonfile: 
        music_json=json.loads(line)
        music_list.append(music_json)
        
music_df = pd.DataFrame(music_list)
music_df  
```

```python
meta_music=[]
i=0
filepath_metamusic = os.path.join("meta_Digital_Music.json")
with open(filepath_metamusic) as jsonfile:
    for line in jsonfile: 
        i+=1
        try: 
            music_json=json.loads(line)
            meta_music.append(music_json)
        except: 
            print(i)
            continue       
            
metamusic_df = pd.DataFrame(meta_music)
metamusic_df
```

```python
price_df=merging_dig_df[['brand','price','overall']]
price_df
```

```python
NaN_df = price_df.replace('',np.nan,regex = True)
price1_df=NaN_df.replace('\n',np.nan,regex=True)
price2_df=price1_df.replace('.a-section.a-spacing-mini{margin-bottom:6px!im...',np.nan,regex=True)
price3_df=price2_df.replace('.a-box-inner{background-color:#fff}#alohaBuyBo...',np.nan,regex=True)
price_df=price3_df.dropna()
price_df=price_df.reset_index(drop=True)
price_df
```
